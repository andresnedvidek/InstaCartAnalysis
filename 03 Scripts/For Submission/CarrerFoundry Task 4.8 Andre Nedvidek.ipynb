## Task 4

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
```

```python
ords_prods_merge.head(15)
```

```python
# Creating a flagged column 
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts()
```

```python
# looking at specific columns to check the data:
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

## Task 5

```python
ords_prods_merge.groupby('loyalty_flag').agg ({'prices':['mean','min', 'max', 'sum']})
```

### The results are quite interesting. It shows that the new customers have the highest purchasing price average followed by regular customers and loyal customers are last. Overall, with the biggest purchasing group being regular customers, it can safely be said that this is where the most revenue is generated.

## Task 6

```python
# Now, I need to create a new column named ave. This will help to group the customers into different spender classes.

ords_prods_merge['spend_ave'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge.head()
```

```python
# Creating a flagged column for spender_typ 
ords_prods_merge.loc[ords_prods_merge['spend_ave'] < 10, 'spender_typ'] = str('low_spender')
```

```python
ords_prods_merge.loc[ords_prods_merge['spend_ave'] >= 10, 'spender_typ'] = str('high_spender')
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['spender_typ'].value_counts()
```

## Task 7 

```python
# Creating a column with the avergae of the days_since_prior_order by customer
ords_prods_merge['frequency_ave'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
ords_prods_merge.head()
```

```python
#creating an additional Flag Column named frequency_user
ords_prods_merge.loc[ords_prods_merge['frequency_ave'] >20, 'customer_freq'] = str('Non-frequent customer')
```

```python
ords_prods_merge.loc[(ords_prods_merge['frequency_ave'] > 10) & (ords_prods_merge['frequency_ave'] <= 20), 'customer_freq'] = str('Regular customer')
```

```python
ords_prods_merge.loc[ords_prods_merge['frequency_ave'] <=10, 'customer_freq'] = str('Frequent customer')
```

```python
ords_prods_merge[['user_id','frequency_ave', 'customer_freq', 'spend_ave','spender_typ']].head(60)
```
